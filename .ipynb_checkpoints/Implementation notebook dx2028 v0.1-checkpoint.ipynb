### The Magnet Set

```python
import numpy as np
import matplotlib.pyplot as plt
import timeit
```

```python
def magnetSet(nx=1024, ny=728, N=100, z_max=30, d_min=0.01):

    x = np.linspace(0.5, 1.25, nx)
    y = np.linspace(-0.2, 0.2, ny)
    X, Y = np.meshgrid(x, y)
    C = X + 1j * Y

    nrows, ncols = C.shape
    results = np.zeros_like(X)
    
    for idx in range(nrows):
        for jdx in range(ncols):
            z = 0 + 0j
            for n in range(N):
                z_old = z
                z = np.sqrt((z ** 2 + C[idx, jdx] - 1) / (2 * z + C[idx, jdx] - 2) )
                if abs(z) > z_max or abs(z-z_old) < d_min:
                    results[idx, jdx] = n
                    break

    return X, Y, results
```

```python
X, Y, fractalSet = magnetSet(N=20, d_min=0.000001)
```

```python
fig = plt.figure(figsize=(16, 8))
ax = fig.gca()
ax.set_aspect('equal')
ax.pcolor(X, Y, fractalSet, shading='auto')
```

```python
%timeit X, Y, fractalSet = magnetSet(N=20, d_min=0.000001)
```

```python
def magnetSetFast(nx=1024, ny=728, N=100, z_max=30, d_min=0.01):

    x = np.linspace(0.5, 1.25, nx)
    y = np.linspace(-0.2, 0.2, ny)
    X, Y = np.meshgrid(x, y)
    C = X + 1j * Y

    nrows, ncols = C.shape
    results = np.zeros_like(X)
    z0 = np.zeros_like(X)
    z = z0 + 0j
    keep = np.full(X.shape, True)
    for n in range(N):
        z_old = z
        z = np.sqrt((z ** 2 + C - 1) / (2 * z + C - 2) )
        mask = (abs(z) > z_max) | (abs(z-z_old) < d_min) & keep
        results[mask] = n
        keep[mask] = False

    return X, Y, results
```

```python
%timeit X, Y, fractalSet = magnetSetFast(N=20, d_min=0.000001)
```
